## Images paths getter

```python
def get_images_paths(path, res, scale=2):
    path_full = path + ('image_SRF_%s/' % (scale if res == 'high' else scale))
    file_ending = '%s_%s.png' % ((scale, 'HR') if res == 'high' else (scale, 'LR'))
    
    return [path_full + file for file in os.listdir(path_full) if file.endswith(file_ending)]
```

```python
get_images_paths('data/BSD 100/', 'low', 3)
```

```python
for path in get_images_paths('data/Set 14/', 'low', 4):
    display(Image.open(path))
```

```python
def count_images(path, res, scale=2):
    return len(get_images_paths(path, res, scale=2))
```
